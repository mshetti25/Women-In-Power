```python
# import the csv with companies female managers
df2="Resources/companies_with_female_managers.csv"

#Read files and store in pandas
df_firms = pd.read_csv(df2)
df_firms.head()
```

```python
#remove additional columns in this case the Indicator Code
df_firms_short = df_firms.drop(["Indicator Code"], axis=1)
df_firms_short
```

```python
# remove colomn years between years 1960 to 2018
df_companies= df_firms_short.drop(df_firms_short.iloc[:,6 :55], axis=1)
df_companies
```

```python
# rename the colomns with shorter names and lower letters
df_companies_rename= df_companies.rename(columns= {
    "Country Name":"country",
    "Country Code":"country_code",
    "new code": "new_code",
    "Indicator Name":"indicator_name",
    "2009":"year_2009",
    "2010":"year_2010",
    "2011":"year_2011",
    "2012":"year_2012",
    "2013":"year_2013",
    "2014":"year_2014",
    "2015":"year_2015",
    "2016":"year_2016",
    "2017":"year_2017",
    "2018":"year_2018",
    "2019":"year_2019",
})
df_companies_rename
```

```python
# change the indicator name column with a shorter name
df_companies_rename["indicator_name"].replace({"Firms with female top manager (% of firms)": "firms with female manager"}, inplace=True)
df_companies_rename
```

```python
# drop NaN values from the new code column---this will drop all the unwanted countries
df_companies_rename2=df_companies_rename.dropna(subset= ["new_code"])
df_companies_rename2
```

```python
# # fill the NaN with "0" value
df_firms_fill= df_companies_rename2.fillna(0)
df_firms_fill
```

```python
# drop the 2020 column
df_firms_final = df_firms_fill.drop(["2020"], axis=1)
df_firms_final.head()
```

```python
# round all the numbers to only 2 decimals
df_firms_final_round =df_firms_final.round(2)
df_firms_final_round.head()
```

```python
#save to csv in the Data folder
df_firms_final_round.to_csv("Data/female_manager.csv" ,index = False, header=True)
```
